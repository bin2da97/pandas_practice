```python
# request
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# headers에 'user-agent'값

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'}
number =[]
name = []
position = []
age = []
nation = []
team = []
value = []
for i in range(1,3):
    # url
    url = f'https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop?ajax=yw1&page={i}'

    #requset.get
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text,'html.parser')
    player_into = soup.find_all('tr', class_=['odd','even'])
    
    for info in player_into:
        player = info.find_all('td')
    
    for info in player_into:
        player = info.find_all('td')
        number.append(player[0].text)
        name.append(player[3].text)
        position.append(player[4].text)
        age.append(player[5].text)
        nation.append(player[6].img['alt'])
        team.append(player[7].img['alt'])
        value.append(player[8].text.strip())
    
    time.sleep(1)
```

```python
# 정보 담을 리스트
```

```python

print(value)
```

```python
df = pd.DataFrame(
        {"number": number,
        "name":name,
        "posiotion":position,
        "age":age,
        "nation":nation,
        "team":team,
        "value":value}
)
df
```

```python
df.to_csv('transfermarkt50.csv',index=False)
```
